```python
import moviepy.editor as mp
import numpy as np
import pandas as pd
import re
import moviepy.editor

#Please use video-file and comment from channel-page
#That project cut video-fragment by 10 seconds after marked timecode in comment.
#If you want to change duration - change ".set_duration(your duration)"

pd.set_option('display.max_rows', None)
not_sorted_df = pd.read_csv("OS_Path/file.csv") # use Your csv-file containing comments



del_useless_col = not_sorted_df.drop(['channel_hyperlink', 'video_hyperlink', 'channeltitle', 'commentid',
                                      'authorchannelurl', 'authordisplayname', 'publishedat', 'totalreplycount', 
                                      'likecount', 'is_parent', 'parent_id', 'api_hash', 'create_timestamp', 
                                      'channelid', 'videoid','reply_count', 'id', 'login', 
                                      'youtube_url', 'channelid.1', 'videoid.1', 'create_timestamp.1', 
                                      'bulkrequest_id'], axis = 1)

del_useless_col.rename(columns={'textoriginal': 'comments'}, inplace = True)

del_useless_col.drop_duplicates(subset="comments", keep=False, inplace=True)


only_comm_with_tc = del_useless_col[del_useless_col.comments.str.contains('\d:\d\d')]
only_comm_with_tc.reset_index(level=False, drop=True, inplace=True, col_level=0, col_fill='')

main_df = only_comm_with_tc



def remove_unneces_char(x: str):
    input_str = x.replace('\n', '')
    input_str = re.sub(r'\d:\d\d', '', input_str)
    return input_str


comment_str = []
comment_str.append(main_df.comments.apply(remove_unneces_char))
comment_list = pd.DataFrame(comment_str)
comm_df = comment_list.T


def take_tc_from_main_df(x: str):
    search_tc = re.search(r'\d:\d\d+', x)
    return search_tc.group(0)

tc_item =[]
tc_item.append(main_df.comments.apply(take_tc_from_main_df))
tc_list = pd.DataFrame(tc_item)

tc_df = tc_list.T
tc_df.columns = ['timecodes']





def tc_to_secs(x: str):
    min_into_secs, secs = re.split(r':', x, maxsplit=1)
    retrieval_secs = (int(min_into_secs) * 60) + int(secs)

    return retrieval_secs

secs_list = []
secs_list.append(tc_df.timecodes.apply(tc_to_secs))
secs_list = pd.DataFrame(secs_list)
secs_df = secs_list.T

    
        
    
parralel_df = main_df
concated_df = pd.concat([comm_df, tc_df], axis=1, join='outer')


main_video = moviepy.editor.VideoFileClip("OS_Path/file.mp4") # Use your video file
main_video_duration = int(main_video.duration)
for i in range (len(secs_df) - 500):
    video_name = 'Kazino joke ' + str(i)
    time_code_n = secs_df.iloc[i, 0]
    comment_n = comm_df.iloc[i, 0]
    if time_code_n + 10 > main_video_duration:
        video_fragment = main_video_duration - time_code_n
        video = mp.VideoFileClip("OS_Path/file.mp4").subclip(time_code_n,time_code_n + video_fragment) # Use your video file
    else:
        video = mp.VideoFileClip("OS_Path/file.mp4").subclip(time_code_n,time_code_n+10) # Use your video file
    if not comment_n:
        txt_clip =  mp.TextClip(video_name,color='white', align='West', fontsize=20, font='Arial', method='caption').set_position(("center", "bottom")).set_duration(10)
    else:
        txt_clip =  mp.TextClip(comment_n,color='white', align='West', fontsize=20, font='Arial', method='caption').set_position(("center", "bottom")).set_duration(10)
    output_video = mp.CompositeVideoClip([video,txt_clip])
    output_video.write_videofile(video_name+".mp4",fps=25)
```
